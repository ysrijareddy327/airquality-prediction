```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    pollution_df[col] = le.fit_transform(pollution_df[col])
    label_encoders[col] = le

# Step 3: Correlation Heatmap
# Compute the correlation matrix
correlation_matrix = pollution_df[numerical_cols + categorical_cols].corr()

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap Between Numerical and Encoded Categorical Data')
plt.show()
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import pandas as pd


# Step 1: Feature Selection and Target Separation
X = pollution_df.drop(columns=['AQI CATEGORY'])  # Features
y = pollution_df['AQI CATEGORY']  # Target

# Step 2: Standard Normalization of Features
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
# Step 3: Split the Data into Train and Test Sets (80/20)
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Step 4: Model Selection and Training
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "Decision Tree": DecisionTreeClassifier(),
    "K-Nearest Neighbors": KNeighborsClassifier(),
}

results = []

for model_name, model in models.items():
    print(f"Training {model_name}...")
    # Train the model
    model.fit(X_train, y_train)
    # Make predictions
    y_pred = model.predict(X_test)

    # Step 5: Evaluation Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    conf_matrix = confusion_matrix(y_test, y_pred)

    # Document results
    results.append({
        "Model": model_name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1
    })

    # Print metrics for each model
    print(f"{model_name} Metrics:")
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
    print(f"Confusion Matrix:\n{conf_matrix}\n")

# Step 6: Summary of Results
results_df = pd.DataFrame(results)
print("\nSummary of Model Performances:")
print(results_df)

# Save the results
with open("results.txt", "w") as file:
    file.write(str(accuracy))


```

```python
# Import necessary libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Step 6: Present Results in a Table
results_df = pd.DataFrame(results)

# Display the results in a tabular format
print("\nSummary of Model Performances:\n")
print(results_df)

# Save the results
results_df.to_csv('/content/kaggle.json', index=False)
print("\nModel results saved")

# Step 7: Visualizing the Metrics
# Set the style
sns.set(style="whitegrid")

# Accuracy Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x="Model", y="Accuracy", palette="viridis")
plt.title("Model Accuracy Comparison", fontsize=16)
plt.ylabel("Accuracy", fontsize=12)
plt.xlabel("Model", fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()

# Precision Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x="Model", y="Precision", palette="coolwarm")
plt.title("Model Precision Comparison", fontsize=16)
plt.ylabel("Precision", fontsize=12)
plt.xlabel("Model", fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()

# Recall Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x="Model", y="Recall", palette="magma")
plt.title("Model Recall Comparison", fontsize=16)
plt.ylabel("Recall", fontsize=12)
plt.xlabel("Model", fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()

# F1 Score Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df, x="Model", y="F1 Score", palette="cividis")
plt.title("Model F1 Score Comparison", fontsize=16)
plt.ylabel("F1 Score", fontsize=12)
plt.xlabel("Model", fontsize=12)
plt.xticks(rotation=45)
plt.ylim(0, 1)
plt.show()
```
